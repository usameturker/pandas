# 4. Data Filtreleme 
### | Uygulama

```python
df = pd.read_csv('developer-survey-2019/survey_results_public.csv', index_col='Respondent')

df.head(5)
```

```python
# Uygulama-1 :

# "Maaş"ı (ConvertedComp) 70000 den yüksek olanları sırasıyla ;
# "Ülke", "Çalıştıkları Programlama Dili", "Maaş" bilgilerini getirelim.

# Maaş : 'ConvertedComp'  ,   Ülke : 'Country'  ,  
# Çalıştıkları Programlama Dili : 'LanguageWorkedWith' 

# Uygun filtreyi yazıp doğru sonucu getiren indexlemeyi data frame'e uygulayalım...
```

```python
# Cevap : 
# filtre ;

my_filt = (df['ConvertedComp'] > 70000)

# Sonucu getiren indexlemenin data frame'e uygulanması ;

df.loc[my_filt, ['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```

```python
# Uygulama-2 :

# Aşağıdaki ülkelerden birinden olanları bulan filtreyi yazalım.
# 'United States', 'India', 'United Kingdom', 'Germany', 'Canada'
# filtreyi data frame'e sadece ülke 'Country' bilgisini getirecek şekilde uygulayalım.
```

```python
# Cevap : 
# filtre ;

country_list = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']

my_filt = df['Country'].isin(country_list)

# Sonucu getiren indexlemenin data frame'e uygulanması ;

df.loc[my_filt, ['Country']]
```

```python
# Uygulama-3:

df['LanguageWorkedWith']
```

```python
# LanguageWorkedWith sütunundaki değerlerin string ve birçok kelimeden
# meydana gelen veriler olduğunu görüyoruz. Bu string veriler içerisinde
# tek bir kelimeyi aramak istersek, örneğin Python bilenleri getirme durumunda
# "LanguageWorkedWith" te Python string'i geçen verileri getirmemiz gerekir.
# Bunun için özel bir method olan str.contains() methodunu kullanacağız.
# Ancak birşeye dikkat etmeliyiz;
# LanguageWorkedWith sütununda veri girişi yapılmamış yani 'NaN' satırlar var.
# Bu satırlarda 'Python' kelimesini aramak istersek hata alırız. 
# Hata almamak için str.contains() methodunda na=False gireceğiz.
# Şimdi uygulayalım :

filt = df['LanguageWorkedWith'].str.contains('Python', na=False)

df.loc[filt, 'LanguageWorkedWith' ]
```

```python

# Aynı örnekte birden fazla string ifade aramak istersek ;

new_filt = (df['LanguageWorkedWith'].str.contains('Python|HTML|Java', na=False))

df.loc[new_filt, 'LanguageWorkedWith' ]
```

```python
country_grp = df.groupby(['Country'])  

# Burada groupby methodu dataframe'i, Country sütununun içerisindeki verilere 
# göre grupladı ve biz bu objeyi country_grp değişkenine atadık.
```

```python
country_grp
```

```python
# Bu şekilde doğrudan görüntülemek istediğinizde
# country_grp isimli değişkende bir "object" tutulduğunu söylüyor.
# Biz bu object'in attribute ve methodları üzerinden ihtiyacımız olan sonuçlara ulaşacağız :

country_grp.get_group('United States')
```

```python
# Sonuca baktığımızda bize Country değeri  United States olan verileri
# dataframe'imizden bir grup olarak getirdiğini görüyoruz.

# Aynı sonucu tabiki filtreleme yaparak da alabiliriz :

filt = ( df['Country'] == 'United States' )
df.loc[filt]
```

```python
# Eğer bu şekilde filtreleme yolunu seçersek,
# herbir ülke için ayrı ayrı filtre yazmamız gerekir.
# Grouping ise bizi bunu otomatik olarak yapmış oluyor.
# ve hangi ülkeyi görmek istiyorsak onu argument olarak 
# get_group('United States') şeklinde girmemiz yeterli oluyor.

# Grouping yapıp istenilen grubu çekerek ilk iki fazı ; 
# Splitting ve Apply Function'i tamamlamış olduk.
# Şimdi Combine results (Sonuçları birleştirme) örneğine geçelim :

# örneğin United States'ten olan katılımcılar arasında en çok kullanılan sosyal medyaları
# görmek istiyoruz diyelim :

filt = ( df['Country'] == 'United States' )

df.loc[filt]['SocialMedia'].value_counts()
```

```python
# Bu işlemi grouping üzerinden ilerleterek her bir ülke için
# görmek istersek ;
# Yukarıda şöyle bir gruplama tanımlamıştık ülkelere göre ;
#     country_grp = df.groupby(['Country']) 
# Şimdi Bu gruplama içerisinde SocialMedia verilerini görmek için :

country_grp['SocialMedia'].value_counts()
```

```python
# İlk Ülke ve son ülkenin Social Media verilerini görebiliyoruz. Arada kalanlar
# özetlenmiş. Ancak Ülkelerin Social Media verileriyle birleştirildiğini görüyoruz.
# Herbir ülkede kullanılan Social media sayıları görülüyor.

country_grp['SocialMedia'].value_counts()
```

```python
# Şimdi bu sonucu ;
# United States'ten olan katılımcılar arasında kullanılan sosyal medyaların
# sayılarını görmek için uyarlarsak ;
# country_grp['SocialMedia'].value_counts()  ifadesinin 
# bize Country değerlerini satır label'ları olarak aldığını görüyoruz.
# o halde bu sonucun üzerinden satır label'ını United States olarak girip 
# istediğimiz sonucu çekebiliriz :

country_grp['SocialMedia'].value_counts().loc['United States']
```

```python
# Groupping niçin daha kullanışlı ?

# her bir kriter için ayrı ayrı filtre yazmak yerine 
# bir kez groupping yaparak istediğimiz kriterle ilgili 
# başka sonuçları da birleştirerek (combine / aggregating) yeni istatistik bilgileri elde edebiliyoruz,
# sonuçları birleştirebiliyoruz.

# Örneğin Rusyadaki katılımcılar arasında kullanılan sosyal medayaların yüzde 
# dağılımlarını görelim :
# normalize=True yapmak yeterli.

country_grp['SocialMedia'].value_counts(normalize=True).loc['Russian Federation']
```

```python
# Ülkelerin maaş değerlerinde ortalama(mean) değerleri görelim :

country_grp['ConvertedComp'].mean()
```

```python
# Ülkelerin maaş değerlerinde median değerleri görelim :

country_grp['ConvertedComp'].median()
```

```python
# Türkiye'deki  median maaşı görmek için :
# country_grp['ConvertedComp'].median() sonucunda bize 
# satır labelları ülkeler olduğuna göre ;

country_grp['ConvertedComp'].median().loc['Turkey']
```

```python
# iki ayrı methodu birleştirip sonuçlarını bir tabloda görmek(birleştirmek)
# için agg() methodunu kullanırız.    agg: aggregation/birleştirme

country_grp['ConvertedComp'].agg( ['median', 'mean'] )
```

```python
# Böylece herbir ülkeden Python bilen kaç kişi var getirmiş olduk.

# Bu sonuçta, Ülkelerin satır label olduklarını görüyoruz.
# öyleyse belirli bir ülkeden - United States'ten Python bilen kaçkişi var ?
# sorusunun cevabı ;

country_grp['LanguageWorkedWith'].apply( lambda x: x.str.contains('Python').sum() ).loc['United States']
```

```python
# şimde yüzde değerlerine ulaşmak için value_counts() methodunu
# burada doğrudan kullanamayız. value_counts ile ulaşacağımız yüzde ancak diğer ülkelere
# olan oranlamanın yüzdesi olur.

# Her bir ülkeden ankete katılanların toplam sayıları; (daha önce bulmuştuk)

country_resp = df['Country'].value_counts()

country_resp
```

```python
# Her bir ülkeden Python bilenlerin toplam sayılarını da
# bir değişkende tutalım ;

country_uses_python = country_grp['LanguageWorkedWith'].apply( lambda x: x.str.contains('Python').sum() )

country_uses_python
```

```python
# Şimdi herbir ülkeden katılanların toplam sayısı ile
# her bir ülkeden Python bilenlerin sayılarını birleştiren bir 
# dataframe oluşturalım ;

# Elimizde şuan  country_resp ve country_uses_python olarak
# isimlendirdiğimiz iki tane Series datası var.

# İkisini birleştirip bir dataframe oluşturalım ;
# birleştirmek için concat() methodunu kullanacağız :
python_df = pd.concat([country_resp, country_uses_python], axis='columns', sort=False)

# sort parametresini false yaptık, alfabetik sorting yapmasın diye, axis'i sütunlar olarak belirledik;
# yanyana sütunlar olarak birleştirmesi için.


python_df    # oluşturduğumuz dataframe'i görelim :
```

```python
# Country sütunu herbir ülkeden katılımcılaarın sayılarını gösteriyor.
# LanguageWorkedWith sütunu bu ülkelerden Python bilenlerin sayısını gösteriyor.

# Şimdi yeni dataframe'imizde ;

# 'Country' sütununun label'ını 'NumRespondents' olarak
# 'LanguageWorkedWith' sütununun label'ını 'NumKnowsPython' olarak değiştirelim.

# sütun labellarını değiştirmek için rename() methodunu kullanalaım :

python_df.rename(columns={'Country': 'NumRespondents', 'LanguageWorkedWith': 'NumKnowsPython'}, inplace=True)
```

```python
python_df 
```

```python
# Şimdi bu dataframe'e herbir ülkeden Python bilenlerin yüzde değerlerini
# ayrı bir sütun olarak ekleyelim :
# yüzde değerlerini tutan sütunun label'ı : 'PctKnowsPython' olacak.

python_df['PctKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents']) * 100

python_df
```

```python

# Python bilenlerin oranı (yüzdesi) en fazla olandan en aza doğru
# ülkeleri sıralamasını (sorting) istersek :

python_df.sort_values(by='PctKnowsPython', ascending=False, inplace=True)

python_df.head(50)    # ilk 50 ülkeyi görelim.
```

```python
# Bu ülkeler içinde Türkiyenin değerlerini getirelim :
# Ülke isimleri satır label'ları olduğuna göre şöyle getirebiliriz ;

python_df.loc['Turkey']
```

```python
na_vals = ['NA', 'Missing']   # Datasetimizde string olarak gelen 'Missing' ve 'NA' girdilerinin gerçek NA'ye dönüşmesi
                              # için önce bu stringleri bir array içerisine aldık.
                              # csv datamızı okurken bu stringlerin NA değerler olarak alınması için na_values 
                              # parametresine bu array'i atayacağız.
df = pd.read_csv('developer-survey-2019/survey_results_public.csv', index_col='Respondent', na_values=na_vals )
schema_df = pd.read_csv('developer-survey-2019/survey_results_schema.csv', index_col='Column')

pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
```

```python
# küçük datasetimizdeki çözümü uygulayalım ve float'a dönüştürüp 
# öyle hesaplama yapmayı deneyelim ;
df['YearsCode'] = df['YearsCode'].astype(float)

# 'Less than 1 year' şeklinde string ifadeler olduğu için hata alığmızı görüyoruz.
# sadece sayı içeren string ifadeler olsaydı böyle bir hata almayacaktık.
# beklemediğimiz bir durumla karşı karşıyayız.
```

```python
# Öncelikle bu sütundaki unique (benzersiz) değerlerin neler olduğuna bakalım ;

df['YearsCode'].unique()
```

```python
# sayılar dışında farklı string ifadeler olduğunu da görüyoruz ;
# 'Less than 1 year' ve 'More than 50 years' ifadeleri bunlar.
# 'Less than 1 year' ifadesi girilen değerleri 0 ile (1'den az olduğu için) değiştirelim :

df['YearsCode'].replace('Less than 1 year', 0, inplace=True)
```

```python
# 'More than 50 years' ifadesini de 51 sayısı ile değiştirelim.
# bu cevabı veren kişi sayısı çok az olacağı için 51 yapmakla 52,53 veya 
# başka bir sayı yapmak hesaplamamızı ciddi şekilde etkilemeyecektir.

df['YearsCode'].replace('More than 50 years', 51, inplace=True)
```

```python
df['YearsCode'].unique()  # sayılar dışında stringler kalmadığını görüyoruz :
```

```python
# Ancak dtype hâlâ object olduğu için mean-median hesabından önce 
# type'mızı floata dönüştüreceğiz. 

df['YearsCode'] = df['YearsCode'].astype(float)
```
